```python
class Task: # Создание объекта - задачи
  def __init__(self,task_id,name,description,status,duration):
    self.__task_id=task_id
    self.__name=name
    self.__description=description
    self.__status=status
    self.__duration=duration
  def get_id(self):
    return self.__task_id
  def get_name(self):
    return self.__name
  def get_description(self):
    return self.__description
  def get_status(self):
    return self.__status
  def get_duration(self):
    return self.__duration

class Project: # Создание объекта - списка задач, реализующего различные методы
  def __init__(self):
    self.tasks=[]
  def add_task(self,task): # Добавление задачи в список
    self.tasks.append(task)
  def remove_task(self,task_id):  # Удаление задачи
    for i, task in enumerate(self.tasks):
      if task.get_id()==task_id:
        del self.tasks[i]
        return True
    return False
  def get_all_tasks(self):  # Получение всех задач
    return self.tasks
  def __iter__(self):  # Итерация
     return iter(self.tasks)

class Analysis:  # Класс анализа задач
  @staticmethod
  def count_tasks(project): # Количество задач в проекте
    return len(project.tasks) if hasattr(project, 'tasks') else 0
  @staticmethod
  def count_tasks_by_status(project, status): # Количество задач в проекте с зад. статусом
    count = 0
    for task in project:
      if task.get_status()==status:
        count+=1
    return count
  @staticmethod
  def find_longest_task(project):  # Поиск задачи с наибольшей длительностью
    m=project.tasks[0]  # Начинаем с первой задачи
    for task in project:
      if task.get_duration() > m.get_duration():
        m=task
    return m


class EnhancedAnalysis(Analysis): # Класс дополнительного анализа
  @staticmethod
  def find_tasks_with_keywords(project, keywords):  # Поиск задач по ключевым словам в имени или описании
    f=[]
    for task in project:
      for i in keywords:
        if (i in task.get_name()) or (i in task.get_description()):
          f.append(task)
          break
    return f



```

```python
task1 = Task(1, "Задача 1", "Описание1", "Выполняется", 1)
task2 = Task(2, "Задача 2", "Описание2", "Выполняется", 8)
task3 = Task(3, "Урок 1", "Описание1", "Завершена", 3)
task4 = Task(4, "Задача 3", "Описание: срочно", "Выполняется", 7)
task5 = Task(5, "Урок 2", "Описание1", "Завершена", 3)
task6 = Task(6, "Задача 4", "Описание: срочно", "Выполняется", 7)
project = Project()
project.add_task(task1)
project.add_task(task2)
project.add_task(task3)
project.add_task(task4)
project.add_task(task5)
project.add_task(task6)
print(f"Всего задач: {len(project.tasks)}")
result = project.remove_task(2)
print(f"Удаление задачи 2: {'Успешно' if result else 'Не найдено'}")
result = project.remove_task(99)
print(f"Удаление задачи 99: {'Успешно' if result else 'Не найдено'}")
print(f"Осталось задач: {len(project.tasks)}")
print("\nОставшиеся задачи:")
for task in project:
    print(f"  ID: {task.get_id()}, Название: {task.get_name()}")
print(f"Всего задач: {Analysis.count_tasks(project)}")
print(f"Всего задач с заданным статусом: {Analysis.count_tasks_by_status(project, 'Завершена')}")
print(f"Задача с самым длительным временем выполнения: {Analysis.find_longest_task(project).get_name()}")
f=EnhancedAnalysis.find_tasks_with_keywords(project, ['Описание1','Урок'])
print("Названия найденных задач по ключевым словам ")
for task in f:
    print(f"{task.get_name()}")
```
