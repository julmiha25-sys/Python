```python
import os
import csv
from datetime import datetime, timedelta

def get_missing_campaign_dates(file_path):
    if not os.path.exists(file_path):
        return

    with open(file_path, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        headers = next(reader)
        s=headers.index('Начальная дата') # Индекс поля

        starts = {datetime.strptime(row[s].strip()[:10],'%Y-%m-%d').date()
            for row in reader if len(row)>s and row[s].strip()} # Множество дат начала запуска

        if not starts:
            return

        min_d, max_d = min(starts), max(starts) # Поиск диапазона дат

        for i in range((max_d-min_d).days + 1):
            date = min_d+timedelta(days=i)
            if date not in starts:
                print(date.strftime('%Y-%m-%d'))
                yield date  # Возвращение объекта date


answers = list(get_missing_campaign_dates('campaign_data.csv'))
```
