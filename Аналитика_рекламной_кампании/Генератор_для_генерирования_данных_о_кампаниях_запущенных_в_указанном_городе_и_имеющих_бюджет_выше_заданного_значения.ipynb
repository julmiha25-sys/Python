```python
import os
import csv

def campaign_generator(file_path, city_name=None, budget=None):
    if not os.path.exists(file_path):
        return

    with open(file_path, 'r', encoding='utf-8') as f:
        reader=csv.reader(f)
        headers=next(reader)
        i,t,p,b,c,d=map(headers.index, ['ID Кампании', 'Тип кампании', 'Платформа', 'Бюджет', 'Город', 'Доход']) # Определение индексов

        campaigns=[] # Список компаний в виде словарей
        for row in reader:
            if len(row)>max(i,t,p,b,c,d):
                try:
                    city=row[c].strip() # Определение городов
                    budg=float(row[b] or 0) # Определение бюджетов
                    if (city_name is None or city==city_name) and (budget is None or budg>budget):
                        income_value = row[d].strip() # Доход как строка
                        # Добавление элемента списка как словаря
                        campaigns.append({'ID Кампании': row[i].strip(),'Тип': row[t].strip(),'Платформа': row[p].strip(),'Доход': income_value if income_value else '0' })
                except (ValueError, IndexError):
                    continue

        campaigns.sort(key=lambda x: x['ID Кампании']) # Сортировка по ID Кампании (как строке)

        for campaign in campaigns:
            yield campaign
```

```python
for campaign in campaign_generator('campaign_data.csv', 'Москва', 48000):
    print(campaign)
```
