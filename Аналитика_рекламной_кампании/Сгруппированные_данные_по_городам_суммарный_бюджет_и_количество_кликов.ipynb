```python
import os
import csv
from collections import defaultdict # Для создания словаря со значениями по умолчанию

def group_campaign_data(file_path):

    if not os.path.exists(file_path):
        return
    with open(file_path, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        headers = next(reader)
        c,k,b = map(lambda x: headers.index(x), ['Город', 'Клики', 'Бюджет']) # Поиск индексов

        mas = defaultdict(lambda: [0.0, 0.0])  # Список кликов и бюджета

        for row in reader:
            if len(row)>max(c,k,b) and row[c].strip():
                try:
                    city = row[c].strip() # Список городов
                    mas[city][0] += (float(row[k]) if row[k].strip() else 0) # Добавление кликов
                    mas[city][1] += float(row[b]) if row[b].strip() else 0 # Добавление бюджета
                except ValueError:
                    continue

        return [{'Город':city,'Количество кликов':clicks,'Суммарный бюджет':int(budget)} for city, (clicks, budget) in sorted(mas.items())] # Сортировка по городам


group_campaign_data('campaign_data.csv')
```
