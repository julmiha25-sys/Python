```python
import os, csv
from collections import defaultdict # Для создания словаря со значениями по умолчанию

def city_conversion(file_path):
    if not os.path.exists(file_path):
        print(f"❌ Файла '{file_path}' не существует")
        return None

    grouped_sum = defaultdict(float)  # Словарь для кампаний

    with open(file_path, 'r', encoding='utf-8') as f:
      reader = csv.reader(f, delimiter=",", quotechar='"') # Создание объекта для построчного чтения csv-файла с определенным разделителем
      headers = next(reader) # Получение строки заголовка

      city_idx = next((i for i, header in enumerate(headers) if header.strip()=='Город'), None) # Индекс столбца Город
      conversion_idx = next((i for i, header in enumerate(headers) if header.strip()=='Конверсия'), None) # Индекс столбца Конверсия

      data = [(row[city_idx].strip(), float(row[conversion_idx].strip())) for row in reader if len(row) > max(city_idx, conversion_idx) # Столбцов хватает
    and row[city_idx].strip() and row[conversion_idx].strip() and row[conversion_idx].strip().replace('.', '', 1).isdigit() # Проверка данных на валидность
] # Список городов с конверсиями

      sum_dict = {} # Словарь сумм
      count_dict = {} # Словарь количеств

      for city, value in data:
        sum_dict[city] = sum_dict.get(city, 0) + value
        count_dict[city] = count_dict.get(city, 0) + 1

      avg_dict = {} # Словарь средних значений конверсии
      for city in sum_dict:
        avg_dict[city]=sum_dict[city]/count_dict[city]


      most_successful_city=max(avg_dict, key=avg_dict.get) # Максимальная конверсия


    return max(avg_dict, key=avg_dict.get)


city_conversion('campaign_data.csv')
```
