```python
import os, csv


def calculate_average_budget(file_path, platform_name):
    if not os.path.exists(file_path):
        return None

    with open(file_path, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        headers = next(reader)

        p,b = map(lambda x: headers.index(x), ['Платформа', 'Бюджет']) # Поиск индексов

        budgets = (float(row[b]) for row in reader if len(row)>max(p,b) and row[p].strip()==platform_name and row[b].strip() )

        budget_list = list(budgets)

    if not budget_list:
      return 0.0

    return round(sum(budget_list)/len(budget_list),2)

calculate_average_budget('campaign_data.csv', 'Google')
```
