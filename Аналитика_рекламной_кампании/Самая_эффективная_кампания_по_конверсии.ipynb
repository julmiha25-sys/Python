```python
import os, csv
from collections import defaultdict # Для создания словаря со значениями по умолчанию

def company_conversion(file_path):
    if not os.path.exists(file_path):
        print(f"❌ Файла '{file_path}' не существует")
        return None

    grouped_sum = defaultdict(float)  # Словарь для кампаний

    with open(file_path, 'r', encoding='utf-8') as f:
      reader = csv.reader(f, delimiter=",", quotechar='"') # Создание объекта для построчного чтения csv-файла с определенным разделителем
      headers = next(reader) # Получение строки заголовка

      company_idx = next((i for i, header in enumerate(headers) if header.strip()=='Тип кампании'), None) # Индекс столбца Тип кампани
      conversion_idx = next((i for i, header in enumerate(headers) if header.strip()=='Конверсия'), None) # Индекс столбца Конверсия

      data = [(row[company_idx].strip(), float(row[conversion_idx].strip())) for row in reader if len(row) > max(company_idx, conversion_idx) # Столбцов хватает
    and row[company_idx].strip() and row[conversion_idx].strip() and row[conversion_idx].strip().replace('.', '', 1).isdigit() # Проверка данных на валидность
] # Список кампаний с конверсиями

      sum_dict = {} # Словарь сумм
      count_dict = {} # Словарь количеств

      for company, value in data:
        sum_dict[company] = sum_dict.get(company, 0) + value
        count_dict[company] = count_dict.get(company, 0) + 1

      avg_dict = {} # Словарь средних значений конверсии
      for company in sum_dict:
        avg_dict[company]=sum_dict[company]/count_dict[company]


      most_successful_campaign_type=max(avg_dict, key=avg_dict.get) # Максимальная конверсия


    return max(avg_dict, key=avg_dict.get)


company_conversion('campaign_data.csv')

```
