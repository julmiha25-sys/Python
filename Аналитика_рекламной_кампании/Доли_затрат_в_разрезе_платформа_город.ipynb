```python
import os, csv
from collections import defaultdict # Для создания словаря со значениями по умолчанию

def calculate_shares_compact(file_path):
    if not os.path.exists(file_path):
        return None

    with open(file_path, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        headers = next(reader)

        p,c,b = map(lambda x: headers.index(x), ['Платформа', 'Город', 'Бюджет']) # Поиск индексов

        data = defaultdict(lambda: defaultdict(float)) # Словарь с генератором
        for row in reader:
            if len(row)>max(p,c,b): # Столбец существует
                data[row[p].strip()][row[c].strip()]+=float(row[b])

    with open('platform_city_results.txt', 'w', encoding='utf-8') as f:
        for platform, cities in data.items():
            total = sum(cities.values())    # Подсчет суммы затрат по городу
            f.write(f"Для группы {platform}:\n" +
                   ''.join(f"- Город: {city}, доля затрат на рекламу: {round(b/total*100,2)}%\n"
                          for city, b in sorted(cities.items(),
                                                key=lambda item: item[1], # Сортировка по бюджету
                                                reverse=True)) )

    with open('platform_city_results.txt', 'r', encoding='utf-8') as f:
      print(f.read())
    return data


result = calculate_shares_compact('campaign_data.csv')
```
