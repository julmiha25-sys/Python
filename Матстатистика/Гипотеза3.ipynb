```python
import pandas as pd
from scipy.stats import *
import scipy
import numpy as np
df1=pd.read_csv('Salary.csv', header=0)
# Удаление пропусков
#df1=df.dropna()
# Преобразование типа
df1.loc[:, 'Опыт работы']=df1['Опыт работы'].astype(int)
# Количество записей - размер выборки
n=len(df1)
# Выборочная дисперсия
disp_=df1['Зарплата'].var()
sigma0=25300
# Критерий статистики критический
betta=0.98
alpha=1-betta
df_=n-1
chi2_=(chi2.ppf(alpha/2, df_), chi2.ppf(1-alpha/2, df_))
print("Критическая область",chi2_)
#Критерий статистики наблюдаемый
chi2_n=(n-1)*disp_/(sigma0**2)
print("Значение критерия статистики наблюдаемого",chi2_n)
if chi2_n>chi2_[0] and chi2_n<chi2_[1]:
  print("Критерий статистики наблюдаемый не лежит в диапазоне критерия статистики критической - Гипотезу не отвергаем")
else:
  print("Критерий статистики наблюдаемый лежит в диапазоне критерия статистики критической - Гипотезу отвергаем")
p_value = 2*min(chi2.cdf(chi2_n, df_), chi2.sf(chi2_n, df_))
print("p-value",p_value)
if p_value < alpha:
  print("p-value < a - Гипотезу отвергаем")
else:
  print("p-value >= a - Гипотезу не отвергаем")
```
