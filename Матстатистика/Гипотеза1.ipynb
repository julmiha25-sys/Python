```python
import pandas as pd
from scipy.stats import *
import scipy
import numpy as np
df1=pd.read_csv('Salary.csv', header=0)
# Ожидаемая средняя зарплата
M0=63500
# Удаление пропусков
#df1=df.dropna()
# Преобразование типа
df1.loc[:, 'Опыт работы']=df1['Опыт работы'].astype(int)
# Количество записей - размер выборки
n=len(df1)
# Среднее выборочное
mean_=df1['Зарплата'].mean()
# Выборочная дисперсия
disp_=df1['Зарплата'].var()
# Критерий статистики наблюдаемый
t_n=(mean_-M0)/((disp_/n)**0.5)
# Критерий статистики критический
alpha=0.05
df_=n-1
t_k=t.ppf(alpha, df_)
print("Критерий статистики критический",t_k,"Критерий статистики наблюдаемый",t_n)
if t_n<t_k:
  print("Критерий статистики наблюдаемый < Критерий статистики критический - Гипотезу отвергаем")
else:
  print("Критерий статистики наблюдаемый >= Критерий статистики критический - Гипотезу не отвергаем")
p_value=t.cdf(t_n, df_)
print("p-value",p_value)
if p_value < alpha:
  print("p-value < a - Гипотезу отвергаем")
else:
  print("p-value >= a - Гипотезу не отвергаем")
```
