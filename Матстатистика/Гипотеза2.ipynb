```python
import pandas as pd
from scipy.stats import *
import scipy
import numpy as np
df1=pd.read_csv('Salary.csv', header=0)
# Удаление пропусков
#df1=df.dropna()
# Преобразование типа
df1.loc[:, 'Опыт работы']=df1['Опыт работы'].astype(int)
# Количество записей - размер выборки
n=len(df1.dropna(subset=['Уровень образования']))
# Определение доли людей с высшим образованием
df_v=df1[df1['Уровень образования'].isin(['бакалавр', 'магистр','доктор наук'])]
p_v=len(df_v)/n
m=len(df_v)
# Проверка использования z-критерия для доли
if n*p_v>=5:
  if n*(1-p_v)>=5:
    print("z-критерий можно использовать")
else:
  print("z-критерий нельзя использовать")
# Критерий статистики критический
alpha=0.1
p0=0.65
q0=1-p0
z_k=norm.ppf(1-alpha)
# Критерий статистики наблюдаемый
z_n=(p_v-p0)/((p0*q0/n)**0.5)
print("Критерий статистики критический",z_k,"Критерий статистики наблюдаемый",z_n)
if z_n>z_k:
  print("Критерий статистики наблюдаемый > Критерий статистики критический - Гипотезу отвергаем")
else:
  print("Критерий статистики наблюдаемый <= Критерий статистики критический - Гипотезу не отвергаем")
p_value=norm.sf(z_n)
print("p-value",p_value)
if p_value < alpha:
  print("p-value < a - Гипотезу отвергаем")
else:
  print("p-value >= a - Гипотезу не отвергаем")
```
