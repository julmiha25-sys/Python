```python
import numpy as np
def solve_linear_equations(A, b):
  res=np.linalg.solve(A,b)
  return res



 # Допустим есть следующая СЛУ
#2x + 3y + z = 10
# x +  y + z =  6
#3x + 2y + 2z = 12

# Матрица коэффициентов будет выглядеть следующим образом:
A = np.array([[2, 3, 1],
              [1, 1, 1],
              [3, 2, 2]])

# Вектор правой части:
b = np.array([10, 6, 12])

solve_linear_equations(A, b)
# [2.96059473e-16 2.00000000e+00 4.00000000e+00]
```
