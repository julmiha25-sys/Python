```python
# XYZ-анализ продаж аптечной сети
import pandas as pd
import numpy as np
df=pd.read_csv('data1.csv',encoding='1251', delimiter=',')
pd.set_option('display.max_columns',25)
pd.set_option('display.float_format','{:.2f}'.format)
# Расчет количества продаваемого товара
df=df.groupby(['DR_Dat','DR_NDrugs']).agg({'DR_Kol':'sum'}).reset_index()
# Расчет количества дней, в течение которых продавался товар
df1=df.groupby('DR_NDrugs').agg({'DR_Dat': 'count'}).reset_index()
# XYZ-анализ только для товаров, которые продавались более 1 дня за весь период
xyz_names=list(df1[df1['DR_Dat'] > 1]['DR_NDrugs'])
df=df[df['DR_NDrugs'].isin(xyz_names)]
# Расчет коэф. вариации (ср.кв.отклонение/среднее арифм.)
df=df.groupby('DR_NDrugs')['DR_Kol'].apply(lambda x: x.std()/x.mean()).reset_index()
df['XYZ']=np.where(df['DR_Kol']<0.1,'X', (np.where(df['DR_Kol']<0.25,'Y','Z')))
df2=df[['DR_NDrugs','DR_Kol','XYZ']]
df2=df2.rename(columns={'DR_Kol': 'K_var'})
df2
```
