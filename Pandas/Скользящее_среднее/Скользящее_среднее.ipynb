```python
import pandas as pd
import numpy as np
coderun=pd.read_csv('itresume-coderun.csv', encoding='1251')
coderun['date']=pd.to_datetime(coderun['created_at'], format="%Y-%m-%d %H:%M:%S.%f").dt.date
df=coderun.copy()
df=df[df['date'] >= pd.to_datetime('2022-01-01').date()]
df=df.groupby('date').agg(problems=('problem_id', 'count')).reset_index()
df = df.reset_index(drop=True)
df['rolling_avg']=df['problems'].rolling(7, center=True).mean().fillna(0).round(2)
print(df)
```
