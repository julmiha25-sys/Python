```python
import os, csv
from collections import defaultdict # Для создания словаря со значениями по умолчанию

def platform_conversion(file_path):
    if not os.path.exists(file_path):
        print(f"❌ Файла '{file_path}' не существует")
        return None

    grouped_sum = defaultdict(float)  # Словарь для платформ

    with open(file_path, 'r', encoding='utf-8') as f:
      reader = csv.reader(f, delimiter=",", quotechar='"') # Создание объекта для построчного чтения csv-файла с определенным разделителем
      headers = next(reader) # Получение строки заголовка

      platform_idx = next((i for i, header in enumerate(headers) if header.strip()=='Платформа'), None) # Индекс столбца Платформа
      conversion_idx = next((i for i, header in enumerate(headers) if header.strip()=='Конверсия'), None) # Индекс столбца Конверсия

      data = [(row[platform_idx].strip(), float(row[conversion_idx].strip())) for row in reader if len(row) > max(platform_idx, conversion_idx) # Столбцов хватает
    and row[platform_idx].strip() and row[conversion_idx].strip() and row[conversion_idx].strip().replace('.', '', 1).isdigit() # Проверка данных на валидность
] # Список платформ с конверсиями

    grouped_sum = {} # Группировка через словарь - платформа и ее общая конверсия
    for platform, value in data:
        grouped_sum[platform]=grouped_sum.get(platform,0)+value  # Добавление значений в

    most_effective_platform=max(grouped_sum, key=grouped_sum.get) # Максимальная конверсия

    return most_effective_platform

platform_conversion('campaign_data.csv')
```
