```python
import csv
from datetime import datetime
from collections import defaultdict
from collections import Counter # Для посчета количества посещений платформы в популярный час

def analyze(file_path) :
    with open(file_path, 'r', encoding='utf-8') as f:
        reader = csv.reader(f, delimiter=';')
        next(reader)
        rows = [row for row in reader]
    if not rows:
        print("Файл пуст")
        return 0.0
    rows.sort(key=lambda x: (x[2], x[0], x[1])) # Сортировка по задачам

    user_state=defaultdict(list)  # Словарь пользователей со списком часа посещения платформы
    #   user      час
    # {'12852': [16, 15, 13, 14], '11716': [21, 19, 20],

    for row in rows:
        date=row[0]
        user=row[1]

        try:
            time=(datetime.strptime(date, '%Y-%m-%d %H:%M:%S.%f')).hour # Перевод даты в час суток
        except Exception as e:
            print(f"Некорректная дата: {e}")
            continue

        if time not in user_state[user]:  # Проверка нет ли этого часа в списке
            user_state[user].append(time) # Добавление в ловарь

    hour_counter=Counter()

    # Подсчет активности по часам
    for hours_list in user_state.values():
        hour_counter.update(hours_list)  # Добавляем все часы из списка
                                         # {16: 58, 17: 53, 13: 52, ...}

    most_common = hour_counter.most_common(1)[0] # Поиск самого популярного часа и его частоты
    res4=most_common[0]  # Самый популярный час
    return res4

res4=analyze('codesubmit.csv')
print(res4)
```
