```python
from datetime import datetime
import json # Для десериализации JSON-строк в объект Python
from collections import Counter # Для посчета количества строк с конкретным пользователем

def get_average_session_count(file_path):

  total=0  # Сумма длительности сессии
  s=[] # Список встречаемости пользователей

  with open(file_path, 'r') as f:
      next(f)  # Пропуск заголовока
      for line in f:
        try:
            user=line.split(',')[0] # Извлечение части, идущей до первой запятой - имени пользователя
            s.append(user) # Добавление пользователя в список пользователей
            total=sum(Counter(s).values()) # Подсчет посещаемости пользователей
            quantity=len(Counter(s)) # Подсчет количества пользователей

        except:
            continue # В случае ошибки - продолжение поиска в другой сессии
  res4=round((total/quantity),2) # Среднее количество заходов пользователя
  print(res4)
  return res4

res4=get_average_session_count('sessions.csv')
```
