```python
from datetime import datetime
import json # Для десериализации JSON-строк в объект Python
from collections import defaultdict

def get_daily_session_count(file_path):
  day_sessions=defaultdict(list) # Словарь: дата - список пользователей (с повторами)

  with open(file_path, 'r') as f:
      next(f)  # Пропуск заголовка
      for line in f:
        try:
            user=line.split(',')[0] # Извлечение части, идущей до первой запятой - имени пользователя
            part=line.split(',', 1)[1].strip() # Извлечение JSON части, идущей после первой запятой
            part=part.replace('""', '"').strip('"') # Исправление двойных кавычек на одинарные и отбрасывание внешних кавычек
            data=json.loads(part) # Десериализация JSON строки в Python объект
                                  # ['2023-05-20 08:08:50', 744, [['page0', '2023-05-20 08:19:39'], ['page1', '2023-05-20 08:09:51'],...] - строка
            if data and len(data)>2 and data[2]: # Проверка структуры объекта: есть данные, минимум 3 элемента, есть посещенные страницы
              start=(datetime.fromisoformat(data[0].replace(' ', 'T'))).date() # Преобразование время начала сессии из строки в datetime с разделителем Т
              day_sessions[start].append(user) # Добавление в словарь данных о дате сессии и пользователе
        except:
            continue # В случае ошибки - продолжение поиска в другой сессии

  total_sessions=0 # Общее количество всех сессий
  total_user_days=0 # Общее количество уникальных пар (день, пользователь)

  for date, users_list in day_sessions.items():
      total_sessions+=len(users_list) # Все сессии в этот день
      total_user_days+=len(set(users_list)) # Уникальные пользователи в этот день

  # Среднее количество сессий на пользователя в день
  res3 = round(total_sessions / total_user_days, 2) if total_user_days > 0 else 0
  print(res3)
  return res3


res3=get_daily_session_count('sessions.csv')
```
