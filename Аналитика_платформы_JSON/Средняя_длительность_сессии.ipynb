```python
from datetime import datetime
import json # Для десериализации JSON-строк в объект Python

def get_session_duration(file_path):

    total=0  # Сумма длительности сессии
    count=0  # Счетчик сессий

    with open(file_path, 'r') as f:
        next(f)  # Пропуск заголовока
        for line in f:
            try:
                part=line.split(',',1)[1].strip() # Извлечение JSON части, идущей после первой запятой
                part=part.replace('""', '"').strip('"') # Исправление двойных кавычек на одинарные и отбрасывание внешних кавычек

                data=json.loads(part) # Десериализация JSON строки в Python объект
                                             # ['2023-05-20 08:08:50', 744, [['page0', '2023-05-20 08:19:39'], ['page1', '2023-05-20 08:09:51'],...] - строка

                if data and len(data)>2 and data[1]: # Проверка структуры объекта: есть данные, минимум 3 элемента, есть длительность сессии
                    duration=data[1]  # Длительность сессии в сек
                    if duration>=0:
                        total+=duration
                        count+=1
            except:
                continue # В случае ошибки - продолжение поиска в другой сессии
    if count:
      res2=round((total/count),2) # Среднее время длительности сессии
    else:
      res2=0
    print(res2)
    return res2

res2=get_session_duration('sessions.csv')
```
