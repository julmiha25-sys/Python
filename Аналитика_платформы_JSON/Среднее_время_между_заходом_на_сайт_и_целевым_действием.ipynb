```python
from datetime import datetime
import json # Для десериализации JSON-строк в объект Python

def get_time_to_conversion(file_path):

    total=0  # Сумма дельт времени
    count=0  # Счетчик конверсией

    with open(file_path, 'r') as f:
        next(f)  # Пропуск заголовока
        for line in f:
            try:
                part=line.split(',', 1)[1].strip() # Извлечение JSON части, идущей после первой запятой
                part=part.replace('""', '"').strip('"') # Исправление двойных кавычек на одинарные и отбрасывание внешних кавычек

                data=json.loads(part) # Десериализация JSON строки в Python объект
                                             # ['2023-05-20 08:08:50', 744, [['page0', '2023-05-20 08:19:39'], ['page1', '2023-05-20 08:09:51'],...] - строка

                if data and len(data)>2 and data[2]: # Проверка структуры объекта: есть данные, минимум 3 элемента, есть посещенные страницы
                    start=datetime.fromisoformat(data[0].replace(' ', 'T')) # Преобразование время начала сессии из строки в datetime с разделителем Т

                    for page_name, page_time in data[2]:
                        if page_name == "conversion":   # Поиск страницы с "conversion" - целевое действие
                            end=datetime.fromisoformat(page_time.replace(' ', 'T')) # Преобразование времени целевого действия из строки в datetime с разделителем Т
                            total+=(end-start).total_seconds() # Накопление суммы дельт времени от первого захода до конверсии в течении 1 сессии
                            count+=1  # Увеличение счетчика конверсий
                            break  # Выход из цикла - интересует первая конверсия
            except:
                continue # В случае ошибки - продолжение поиска в другой сессии
    if count:
      res=round((total/count),2) # Среднее время конверсии
    else:
      res=0
    print(res)
    return res

res=get_time_to_conversion('sessions.csv')
```
